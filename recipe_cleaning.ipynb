```python
import pandas as pd
import csv

# Step 1: Load the original CSV (adjust path if needed)
df = pd.read_csv('13k_recipes.csv', encoding='ISO-8859-1')
```

```python
# Step 2: Define a function to clean each text field

def escape_quotes_and_clean(text):
    text = str(text)
    text = text.replace('"', '""')  # Escape double quotes for CSV format
    text = text.replace('\r', ' ').replace('\n', ' ')  # Remove line breaks
    text = text.encode('ascii', 'ignore').decode('ascii')  # Remove weird characters
    return text
```

```python
df.head()
```

```python
# Step 3: Apply cleaning to all text columns
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].apply(escape_quotes_and_clean)
```

```python
df.head()
```

```python
# Step 4: Save as a fully quoted, BigQuery-safe CSV
df.to_csv('recipes_13k_bigquery_safe.csv', index=False, quoting=csv.QUOTE_ALL)
```
